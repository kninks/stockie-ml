```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stock_data = yf.download("SCB.BK", period="6mo", auto_adjust=True) # show the adjusted close prices
print(stock_data)
```

```python
stock_data['Close'].plot()
plt.title("SCB Stock Prices")
plt.show()
```

```python
from matplotlib import pyplot
import pandas as pd
from IPython.core.display import display, HTML

display(HTML("<style>.container { width:90% !important; }</style>"))
pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

# function  preprocess  time series 
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True,feat_name=None):
    
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [f'{feat_name[j]}(t-{i})' for j in range(n_vars)]
    
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [f'{feat_name[j]}(t)' for j in range(n_vars)]
        else:
            names += [f'{feat_name[j]}(t+{i})' for j in range(n_vars)]
    
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg

#  list  DataFrame  drop column 'Close'  column 'Adj. Close' 
# stock_data[0] = stock_data[0].drop('Close',axis=1)
# dataset = stock_data[0]
values = stock_data.values
# values = values.astype('float32')

#  scaling  min max scaling  scale  column  [0,1]  min max  column  rescale 
min_dict = dict()
max_dict = dict()
for col in stock_data.columns:
  min_dict[col] = stock_data[col].min()
  max_dict[col] = stock_data[col].max()
  stock_data[col] = (stock_data[col] - stock_data[col].min())/(stock_data[col].max()-stock_data[col].min())
  
#  function  preprocess   drop column 
reframed = series_to_supervised(stock_data.values, 30, 3,feat_name=stock_data.columns)
# reframed.drop(['High(t)','High(t+1)','High(t+2)','Low(t)','Low(t+1)','Low(t+2)','Open(t)','Open(t+1)','Open(t+2)','Volume(t)','Volume(t+1)',\
#                'Volume(t+2)'],axis=1,inplace=True)
```

```python
#  train & test split
values = reframed.values
n_train_percent = 0.9
split = int(values.shape[0]*0.9)
train = values[:split, :]
test = values[split:, :]

#  datetime  train & test
date_reframed = stock_data.index[reframed.index]
date_train = date_reframed[:split]
date_test = date_reframed[split:]

# split feature  target
train_X, train_y = train[:, :-3], train[:, -3:]
test_X, test_y = test[:, :-3], test[:, -3:]

# reshape input  3D (samples, timesteps, features)
train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

#  LSTM network  library Keras  hidden layer(50)  shape  input(#rows x #features)
model = Sequential()
model.add(LSTM(50, input_shape=(train_X.shape[1], train_X.shape[2])))
#  hidden layer  output   forecast (3) 
model.add(Dense(3))
model.compile(loss='mae', optimizer='adam')

# train   log
history = model.fit(train_X, train_y, epochs=100, batch_size=72, validation_data=(test_X, test_y), verbose=2, shuffle=False)

#  train&test loss
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```
